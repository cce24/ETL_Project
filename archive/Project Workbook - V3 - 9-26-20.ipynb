```python
# Dependencies
import pandas as pd
from sqlalchemy import create_engine
from string import digits
import numpy as np
```

```python
#Part 1. Extract
csv1 = "data/DataAnalyst.csv"
csv2 = "data/DataScientist.csv"
df1 = pd.read_csv(csv1)
df2 = pd.read_csv(csv2)
df1.head()
df2.head()
```

```python
df1 = df1.drop(columns=['Unnamed: 0'])
```

```python
df2 = df2.drop(columns=['Unnamed: 0'])
del df2['index']
```

```python
df2.head()
```

```python
df3 = pd.merge(df1, df2, how="outer")
df3.head()
len(df3['Job Title'])
```

```python
#Part 2. Transform
df3 = df3.drop(columns=['Rating', 'Competitors' , 'Easy Apply', 'Founded'])
```

```python
df3.head()
```

```python
df3['Salary Estimate'] = df3['Salary Estimate'].str.replace('K','')
df3['Salary Estimate'] = df3['Salary Estimate'].str.replace('(Glassdoor est.)','')
df3['Salary Estimate'] = df3['Salary Estimate'].str.replace('Employer est.','')
df3['Salary Estimate'] = df3['Salary Estimate'].str.replace('(','')
df3['Salary Estimate'] = df3['Salary Estimate'].str.replace(')','')
df3['Salary Estimate'] = df3['Salary Estimate'].str.replace('$','')
```

```python
df3.head()
```

```python
new = df3['Salary Estimate'].str.split("-", n = 1, expand = True) 
df3['Salary Bottom']= new[0]
df3['Salary Top']= new[1]
df3.head()
```

```python
df3['Revenue'] = df3['Revenue'].str.replace('Unknown / Non-Applicable','0')
df3
```

```python
df3['Revenue - Units'] = df3['Revenue'].str.slice(start=-13)
df3
```

```python
df3['Revenue'] = df3['Revenue'].str.replace('billion','')
df3['Revenue'] = df3['Revenue'].str.replace('million','')
df3['Revenue'] = df3['Revenue'].str.replace('USD','')
df3['Revenue'] = df3['Revenue'].str.replace('(','')
df3['Revenue'] = df3['Revenue'].str.replace(')','')
df3['Revenue'] = df3['Revenue'].str.replace('$','')
df3.head()
```

```python
new2 = df3['Revenue'].str.split("to", n = 1, expand = True) 
df3['Revenue Bottom']= new2[0]
df3['Revenue Top']= new2[1]
df3.head()
```

```python
df4 = df3[['Job Title', 'Job Description', 'Salary Bottom',\
           'Revenue', \
           'Salary Top', 'Company Name', 'Revenue Bottom',\
           'Revenue Top', 'Revenue - Units', \
           'Location', 'Headquarters', 'Size',\
           'Type of ownership', 'Industry', 'Sector']]
```

```python
df6 = df4[df4['Revenue Bottom']!='10+  ']
df6 = df6[df6['Revenue Bottom']!='Less than 1  ']
df6.head()
```

```python
# df4['Revenue Bottom'] = df4['Revenue Bottom'].str.replace('Unknown / Non-Applicable','0')
# df4.head()
```

```python
df6['Revenue Top'] = df6['Revenue Top'].fillna(0)
df6['Revenue Bottom'] = df6['Revenue Bottom'].fillna(0)
df6['Salary Top'] = df6['Salary Top'].fillna(0)
df6['Salary Bottom'] = df6['Salary Bottom'].fillna(0)
df6.head()
```

```python
df6 = df6[~df6["Salary Top"].str.contains('Per', na=False)]
```

```python
df6 = df6[df6["Headquarters"]!= '-1']
df6['Revenue Top'] = df6['Revenue Top'].replace('1','1000')
df6.head()
```

```python
df6['Revenue Bottom'] = df6['Revenue Bottom'].replace(' ','')
df6['Revenue Top'] = df6['Revenue Top'].replace(' ','')
df6['Salary Bottom'] = df6['Salary Bottom'].replace(' ','')
df6['Salary Bottom'] = df6['Salary Bottom'].replace(' ','')
```

```python
df6.dtypes
```

```python
df6['Revenue Top'] = df6['Revenue Top'].astype(int)
df6['Revenue Bottom'] = df6['Revenue Bottom'].astype(int)
df6['Salary Top'] = df6['Salary Top'].astype(int)
df6['Salary Bottom'] = df6['Salary Bottom'].astype(int)
```

```python
df6.dtypes
```

```python
df6['Revenue Midpoint'] = df6[['Revenue Bottom','Revenue Top']].mean(axis = 1)
df6['Salary Midpoint'] = df6[['Salary Bottom','Salary Top']].mean(axis = 1)
df6.head()
```

```python
df6['Revenue Top'] = df6['Revenue Top'].astype(str)
df6['Revenue Bottom'] = df6['Revenue Bottom'].astype(str)
df6['Salary Top'] = df6['Salary Top'].astype(str)
df6['Salary Bottom'] = df6['Salary Bottom'].astype(str)
# df6['Revenue Midpoint'] = df6['Revenue Midpoint'].astype(str)
```

```python
df6['Salary Units'] = ('Thousands')
```

```python
df6.dtypes
```

```python
df7 = df4[(df4['Revenue Bottom']=='10+  ') | (df4['Revenue Bottom']=='Less than 1  ')]
df7.head()
```

```python
df6 = pd.merge(df6, df7, how="outer")
df6.head()
```

```python
df8 = df6[['Job Title', 'Job Description', 'Salary Bottom',\
           'Salary Top', 'Salary Midpoint', 'Salary Units',\
            'Company Name', 'Revenue','Revenue Bottom',\
           'Revenue Top', 'Revenue - Units', \
           'Location', 'Headquarters', 'Size',\
           'Type of ownership', 'Industry', 'Sector']]
```

```python
df8.to_csv('df8.csv')
```
